### 2.3 Random graphs

Nodes and edges are generated at random. 

<b> Erdos-Renyi model: </b> characterized by 2 parameters, $n$ (number of nodes) and $p$ (probability there's an edge between any two nodes). There's a rapid transition at a particular $p$ value ($p*$). This critical value is $p* = ln \ n/n$. A random graph, $G(n,p)$, is unlikely to be connected (there's a path from every node to every other node) if $p < p*$ and vice versa if $p > p*$.

### 2.4 Generate graphs

Generate a complete graph.

```python
# This generator function takes a list of nodes & 
# enumerates all distinct pairs
def all_pairs(nodes):
    for i, u in enumerate(nodes):
        for j, v in enumerate(nodes):
            if i > j:
                yield u, v
L = [1,2,3,4]
print(all_pairs(L))
```

```python
# Another way with iterables: use itertools
from itertools import combinations

def all_pairs_2(nodes):
    pairs = [",".join(map(str, comb)) for comb in combinations(nodes, 2)]
    return pairs

L = [1,2,3,4]
print(all_pairs_2(L))
```

```python
# Construct a complete graph using all_pairs
def make_complete_graph(n):
    G =  nx.Graph()
    nodes = range(n)
    G.add_nodes_from(nodes)
    G.add_edges_from(all_pairs(nodes))
    return G
```

```python
complete = make_complete_graph(10)
nx.draw_circular(complete,
                node_color = 'g', 
                node_size = 1000,
                with_labels = True)
plt.show()
```

### 2.5 Connected graphs

Check if a graph is connected. Start at any node and check whether we can reach all other nodes. If we can reach node, $v$, then we can reach any of the neighbors of $v$.

```python
# Check neighbors of a node in the "complete" graph above
complete.neighbors(0)
```

```python
# Start any node, s
# Mark s as seen, the mark its neighbors and the neighbors' neighbors
# If all nodes are seen, graph is connected

# Fx returns a set of nodes that can be reached from start
def reachable_nodes(G, start):
    seen = set()
    stack = [start]    # keep track of nodes found
    
    # Each time through loop:
    # 1. remove a node from stack
    # 2. if node has been seen, go back to #1
    # 3. else, add node to seen and its neigbors to stack
    # When stack is empty, can't reach any more nodes, so return seen
    
    while stack:
        node = stack.pop()
        if node not in seen:
            seen.add(node)
            stack.extend(G.neighbors(node))
    return seen 
```

```python
reachable_nodes(complete, 9)
```

```python
# Fx chooses a starting node from nodes in the graph
def is_connected(G):
    start = list(G)[0]    # choose 1st node of graph
    reachable = reachable_nodes(G, start)    # get set of nodes reachable from start node
    
    # If length of list = length of graph's nodes list, graph is connected 
    return len(reachable) == len(G)
    
```

```python
is_connected(complete)
```

### 2.6 Generate ER graphs

```python
from numpy.random import random
```

```python
# Randomly determine if p > some random number between 0 - 1
def flip(p):
    return random() < p    # random() generates random int between 0 - 1

flip(0.4)
```

```python
# This generator fx enumerates all possible edges
# Uses helper fx, flip, to choose which should be added to graph
def random_pairs(nodes, p):
    for i, u in enumerate(nodes):
        for j, v in enumerate(nodes):
            if i > j and flip(p):
                yield u, v
```

```python
# Generate and return ER graph G(n,p)
# Same as make_complete_graph but uses random_pairs instead of all_pairs
def make_random_graph(n, p):
    G = nx.Graph()
    nodes = range(n)
    G.add_nodes_from(nodes)
    G.add_edges_from(random_pairs(nodes, p))
    return G
```

```python
random_graph = make_random_graph(10, 0.3)
nx.draw_circular(random_graph, 
                 node_color="b", 
                 node_size=1000, 
                 with_labels=True)
plt.show()
```

### 2.7 Probability of connectivity

For given values of n and p, we calculate the probability that G(n,p) is connected. Generate a large number of random graphs and count how many are connected. 

```python
def prob_connected(n, p, iters=100):
    count = 0
    for i in range(iters):
        random_graph = make_random_graph(n, p)
        if is_connected(random_graph):
            count += 1
    return count/iters
```

```python
prob_connected(10, 0.3, iters=1000)
```

Out of 10,000 ER graphs with these parameters, ~65% are connected.
